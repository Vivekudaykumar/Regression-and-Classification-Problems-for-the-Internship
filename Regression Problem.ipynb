
<h2>Import Libraries</h2>

```python
#Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

<h2> Get the data </h2>

```python
path=r"C:\Users\VIK\Downloads\dataset.txt"
df=pd.read_csv(path)
df.head()
```

```python
# I have taken X as all the features except Target
X=df.iloc[:,:9]
# I have taken y as Target as it is the column we need to predict
y=df.iloc[:,9:]
```

```python
df.shape
```

<h2>Train Test Split</h2>

```python
# We need to import  train_test_split and split training data and testing data. 
# Here, testing data set is 20% of the whole data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

<h2> Using Linear Regression Algortihm </h2>

```python
# I have used Linear Regression Algorithm
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python

lm.fit(X_train,y_train)
```

```python
# I have stored all the predictions in a variable called y_pred
y_pred=lm.predict(X_test)
```

<h2> Predictions and Evaluations </h2>

```python
# We need to evaluate the accuracy of the model, so I have imported r2 score.
# We find out that the model's accuracy is 81% 

from sklearn.metrics import r2_score
r2_score(y_test, y_pred,multioutput='variance_weighted')
```
